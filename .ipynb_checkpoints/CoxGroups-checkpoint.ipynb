```python
# import stuffs
%matplotlib inline
from datasets import get_mayo, get_veteran, get_lung, get_breasttrn
import numpy as np
from pyplotthemes import get_savefig, classictheme as plt
plt.latex = True
```

```python
d = get_lung(prints=True, norm_in=False, norm_out=False)
# Example: d.iloc[:, :2] for times, events
d
```

```python
from lifelines.estimation import CoxPHFitter, KaplanMeierFitter
from lifelines.plotting import add_at_risk_counts

cox = CoxPHFitter()
cox.fit(d, d.columns[0], d.columns[1])
preds = cox.predict_partial_hazard(d.iloc[:, 2:]).values.ravel()

high = (preds < np.percentile(preds, 25))
low = (preds > np.percentile(preds, 75))
mid = ~high & ~low


plt.figure()

fitters = []
for l, g in zip(['Cox Low', 'Cox Mid', 'Cox High'], [low, mid, high]):
    kmf = KaplanMeierFitter()
    kmf.fit(d.iloc[g, 0], d.iloc[g, 1], label=l)
    kmf.plot(ax=plt.gca())
    fitters.append(kmf)
   
plt.legend(loc='best', framealpha=0.5)
add_at_risk_counts(*fitters)
plt.ylim((0, 1))
```
