```python
# import stuffs
%matplotlib inline
from datasets import get_mayo, get_veteran, get_lung, get_breasttrn
import numpy as np
from pyplotthemes import get_savefig, classictheme as plt
plt.latex = True

from pysurvival.rpart import RPartModel

```

```python
d = get_mayo(prints=True, norm_in=True, norm_out=False)
# Example: d.iloc[:, :2] for times, events
d
```

```python
from lifelines.estimation import KaplanMeierFitter
from lifelines.plotting import add_at_risk_counts
from lifelines.estimation import median_survival_times


rpart = RPartModel(minbucket=10)
rpart.fit(d, d.columns[0], d.columns[1])
preds = rpart.predict(d)

plt.figure()

fitters = []
#for l, g in zip(['Cox High', 'Cox Mid', 'Cox Low'], [high, mid, low]):
for l, g in enumerate(sorted(np.unique(preds))):
    kmf = KaplanMeierFitter()
    kmf.fit(d.iloc[preds==g, 0], d.iloc[preds==g, 1], label=str(l))
    kmf.plot(ax=plt.gca())
    fitters.append(kmf)
    
    print("End survival rate for", l, ":",kmf.survival_function_.iloc[-1, 0])
    #if kmf.survival_function_.iloc[-1, 0] <= 0.5:
    #    print("Median survival for", l, ":",
    #          median_survival_times(kmf.survival_function_))
   
plt.legend(loc='best', framealpha=0.5)
add_at_risk_counts(*fitters)
plt.ylim((0, 1))
```

```python
classes = np.unique(preds)
# Just to be safe
classes.sort()
print(classes)

# Need to group them
low = np.zeros_like(preds, dtype=bool)
high = np.zeros_like(preds, dtype=bool)

high_lim = int(0.25 * len(preds))
low_lim = int(0.25 * len(preds))

# Start with high risk, then low risk
for g in classes:
    # Append to group
    high |= (preds == g)
    print(np.sum(high))
    # Go until the group is a quartile big
    if np.sum(high) >= high_lim:
        break
        
for g in reversed(classes):
    # Append to group
    low |= (preds == g)
    print(np.sum(low))
    # Go until the group is a quartile big
    if np.sum(low) >= low_lim:
        break
        
# Mid is the rest
mid = ~high & ~low


print(np.sum(mid))

print(low.sum() + mid.sum() + high.sum())
```

```python
# Plot the new groups
plt.figure()

fitters = []
for l, g in zip(['RPart High', 'RPart Mid', 'RPart Low'], [high, mid, low]):
    kmf = KaplanMeierFitter()
    kmf.fit(d.iloc[g, 0], d.iloc[g, 1], label=l)
    kmf.plot(ax=plt.gca())
    fitters.append(kmf)
    
    print("End survival rate for", l, ":",kmf.survival_function_.iloc[-1, 0])
    if kmf.survival_function_.iloc[-1, 0] <= 0.5:
        print("Median survival for", l, ":",
              median_survival_times(kmf.survival_function_))
   
plt.legend(loc='best', framealpha=0.5)
add_at_risk_counts(*fitters)
plt.ylim((0, 1))
```
